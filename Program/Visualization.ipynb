```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from channel import Channel
```

```python
N = 1001

x = np.linspace (0,1,N)
y = np.linspace (0,1,N)

X, Y = np.meshgrid (x, y)

w = np.array ([[2,1],[1,2]]) / 3
ch = Channel (w)

Z = np.zeros_like (X)
for col in range (N):
    for row in range (N):
        p = np.zeros ((2,))
        p[0], p[1] = X[col,row], Y[col,row]
        Z[col,row] = ch.nextProb (p)[0]
        if np.isnan (Z[col,row]):
            Z[col,row] = 0

```

```python

fig = plt.figure (figsize=(10,5))
ax = fig.add_subplot (121)
surf = ax.contourf (X,Y,Z,np.arange(0,1,0.001))
plt.colorbar (surf,ax=ax)

ax2 = fig.add_subplot (122)
Z2 = np.ones_like (Z) - Z
step = 200
ax2.quiver (X[::step,::step], Y[::step,::step], Z[::step,::step]-X[::step,::step], Z2[::step,::step]-Y[::step,::step], scale=1)

plt.show ()
```
