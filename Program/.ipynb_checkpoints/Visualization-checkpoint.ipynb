```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from channel import Channel
```

```python

x = np.linspace (0,1,1000)
y = np.linspace (0,1,1000)

X, Y = np.meshgrid (x, y)

w = np.array ([[2,1],[1,2]]) / 3
ch = Channel (w)

Z = np.zeros_like (X)
for col in range (1000):
    for row in range (1000):
        p = np.zeros ((2,))
        p[0], p[1] = X[col,row], Y[col,row]
        Z[col,row] = ch.nextProb (p)[0]
        if np.isnan (Z[col,row]):
            Z[col,row] = 0

```

```python

fig = plt.figure (figsize=(10,5))
ax = fig.add_subplot (121)
surf = ax.contourf (X,Y,Z,np.arange(0,1,0.001))
plt.colorbar (surf,ax=ax)

ax2 = fig.add_subplot (122)
Z2 = np.ones_like (Z) - Z
ax2.quiver (X[::999], Y[::999], Z[::999]-X[::999], Z2[::999]-Y[::999], angles='xy', scale_units='xy', scale=1, linewidth=.5)

plt.show ()
```

```python
uu = np.linspace (0,1,3)
vv = np.linspace (0,1,3)
UU, VV = np.meshgrid (uu, vv)

plt.quiver (UU, VV, 0.3*np.ones_like(UU), 0.2*np.ones_like(VV), scale=1,linewidth=10)
```
